# DASK
- 대용량 데이터를 빠르게 처리하기 위한 방법
- 기존의 방식은 메모리에 모든 데이터 다 올려두는 방식
- DASK는 데이터에서 명령한 내용만 가져오는 방식
- 데이터량이 폭발적으로 많은 빅데이터에선 이 방식이 유리

```python
import dask.dataframe as dd
```

```python
df3 = dd.read_csv("data/crime.csv", dtype=str, error_bad_lines=False, warn_bad_lines=False)
df3
```

```python
df3.head()
```

```python
df3.tail()
```

```python
df3.count().compute()
```

```python
from dask.diagnostics import ProgressBar
pbar = ProgressBar()
pbar.register()
```

```python
%%time
df3.count().compute()
```

```python
%%time
df3.count().compute(scheduler="processes", num_workers=8)
```
