# seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname='/Users/hoon/Library/Fonts/NanumSquareRegular.ttf').get_name()
plt.rc("font", family=font_name)
import matplotlib as mpl
mpl.rcParams["axes.unicode_minus"] = False
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
fig = plt.figure() # 배경 만들기
ax = fig.add_subplot(1, 1, 1) #그래프를 몇개 그릴 것인가? (1행 1열 1번째) 0번째, 1번째, 2번째로 시작함
ax.hist(tips["total_bill"])
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("total bill")
```

```python
# 위에 그래프를 seaborn으로 해봅시다.
ax = sns.distplot(tips['total_bill'])
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("total bill")

# 분포곡선을 없애고 싶으면?
# ax = sns.displot(tips['total_bill'])
```

```python
# 선만 남기고 싶으면?
# displot(df['columns'], hist='False)
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("total bill")
```

```python
# 막대만 남기고 싶으면?
# displot(df['columns'], kde='False')
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("total bill")
```

```python
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("total bill")
```

```python
#### 요일별로 인원 수 파악

# 막대그래프, 이유는 인원 수가 빈도수이기 때문.
# 연속변수이면 박스플롯, 이산변수이면 바플롯

graph = sns.countplot("day", data=tips)
```

```python
#회귀선과 산포도를 한번에 보여주는 그래프 regplot
#R보다 편하게~
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
#회귀선 없이 그래프 그리고 싶으면?
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
# 산포도와 히스토그램을 한번에 나타내는 그래프. jointplot
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# 밀도를 나타내는 kde. 이를 이용한 plot이 있다.
sns.kdeplot(x='total_bill', y='tip', data=tips, shade=True)
```

```python
sns.barplot(x='time', y='total_bill', data=tips)
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
```

```python
sns.violinplot(x='time', y='total_bill', data=tips)
```

```python
#regplot보다 lmplot을 더 많이 쓴다. 이유는 hue 속성 때문
#성별에 따른 tip과 total_bill의 관계를 알고 싶다면?
#regplot은 불가능. lmplot은 hue='sex'로 ... :)

sns.lmplot(x='total_bill', y='tip', hue='sex', markers=["o", "x"], data=tips)
```
